```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
series = pd.read_csv('AirPassengers.csv', header=0, index_col=0, parse_dates=True,
squeeze=True)
```

```python
series
```

```python
plt.figure(1)
```

```python
# line plot
plt.subplot(211)
plt.plot(series)

# histogram
plt.subplot(212)
plt.hist(series)
plt.show()
```

```python
# Square Root Transform
```

```python
# contrive a quadratic time series

series1 = [i**2 for i in range(1,100)]
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(series1)
# histogram
plt.subplot(212)
plt.hist(series1)
plt.show()
```

```python
# square root transform a contrived quadratic time series

from numpy import sqrt
series2 = [i**2 for i in range(1,100)]
# sqrt transform
transform = series2 = sqrt(series2)
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(transform)
# histogram
plt.subplot(212)
plt.hist(transform)
plt.show()
```

```python
import numpy as np
```

```python
dataframe = pd.DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = np.sqrt(dataframe['passengers'])
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(dataframe['passengers'])
# histogram
plt.subplot(212)
plt.hist(dataframe['passengers'])
plt.show()
```

```python
# Log transform
```

```python
from math import exp
series2 = [exp(i) for i in range(1,100)]
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(series2)
# histogram
plt.subplot(212)
plt.hist(series2)
```

```python
import numpy as np
```

```python
transform = np.log(series2)
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(transform)
# histogram
plt.subplot(212)
plt.hist(transform)
plt.show()
```

```python
dataframe = pd.DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = np.log(dataframe['passengers'])
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(dataframe['passengers'])
# histogram
plt.subplot(212)
plt.hist(dataframe['passengers'])
plt.show()
```

```python
# Box cox transform
```

```python
from scipy.stats import boxcox
```

```python
dataframe = pd.DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = boxcox(dataframe['passengers'], lmbda=0.0)
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(dataframe['passengers'])
# histogram
plt.subplot(212)
plt.hist(dataframe['passengers'])
plt.show()
```

```python
dataframe = pd.DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'], lam = boxcox(dataframe['passengers'])
print('Lambda: %f' % lam)
plt.figure(1)
# line plot
plt.subplot(211)
plt.plot(dataframe['passengers'])
# histogram
plt.subplot(212)
plt.hist(dataframe['passengers'])
plt.show()
```
